## Understand fraud

In legal dictionary, fraud is "when a person deliberately practices deception in order to gain something unlawfully or unfairly." In most states, fraud can be either a civil or a criminal wrong in most states. 

Fraud can be either hard fraud or soft fraud. <b>Hard fraud</b> occurs when someone deliberately plans a loss, such as an auto collision, auto theft, or house fire in order to claim payment for damages. Criminal rings are sometimes involved in hard fraud schemes. <b>Soft fraud</b> is far more common than hard fraud. Soft fraud is systematic abuse. For example, in an auto collision claim a claimant may claim more damage than actually occured. In a medical insurance claim an individual or an physician may exaggerate existing conditions that require overtreatment or result in the abuse of medical resources.

## Understand current fraud trends

### In credit cards

If you use a credit card in the U.S., you may already received a new card with a chip technology. This is because the card business has shifted to EMV officially. EMV stands for "Europay, Mastercard and Visa". It is a global standard for cards equipped with computer chips to authenticate chip-card transactions. These new cards are introduced to improve payment security, making it more difficult for fraudsters to counterfeit cards. What's the improvement? The traditional magnetic-stripe cards store unchanging personally identifiable information (PII). Frausters can counterfeit the cards easily. An EMV card is different. It create a uniqe transaction code for every transaction. Even if a hacker stole the chip information from a point of sale, he still cannot use the car again because the transaction code cannot be used again.

Although Card-Present fraud is controlled, Card-Not-Present (CNP) fraud has been rising. Frausters use stolen cards to make online purchases rather than in-store transactions. In the UK, online fraud rose 79 percent in the first three years after EMV became the new standard. Canada and Australia also saw increases in online fraud after they transitioned to EMV cards. Also, while the traditional magnetic-stripe cards are not completed phased out, fraudsters continue to purchase stolen PII from the magnetic-stripe cards, and then apply for brand-new EMV cards.

e-commerce frauds will continue. Customers like the convenience of online ordering with in-store pickup, and same-day shipping, likewise fraudsters. Once a fraudster finds same-day shipping success with a particular retailer, he can target the retailer repeatedly.

## Why does fraud occur?

In order to deter fraud, it will be helpful to understand how and why people commit fraud. 
Knowing the “how” and "why" helps managers and business owners create policies and design internal controls to reduce the occurrence of fraud. 

<img src="images/fraud_triangle.jpg" align = "right" alt="Drawing" style="width: 200px;"/>

There are three conditions that cause fraud to occus: <b>motivation</b>, <b>opportunity</b>, and <b>rationalization</b>. Suppose there is an altruistic physician with a patient who is unable to meet the payment for a required treatment. The "pressure" may be that the ultra-benevolent physian feels compassion to cover the bill. The "opportunity" can be that the physician covers the expense by submitting false claims. And the "rationalization" is the doctor feels justified by helping the patient. However, it is still a fraud. Likewise, suppose an employee is in charge of vendor selection. The "pressure" may be that the employee living beyond the means and facing a personal financial crisis. The "opportunity" is the employees having access to assets and information that allows them to both commit and conceal fraud. Th employee sets up a phantom vendor account and creates false invoices to collect payments to him. The employee may "rationalize" his action by thinking he would return the funds in a later time, or he has been conributing greatly to the company, etc.

## Best practices in anti-fraud strategy

Many organisations are building formal systems and procedures in place to prevent, detect and respond to fraud. [Kroll](https://www.kroll.com/en-us/default.aspx) suveyed 1,265 executives worldwide and [reports](https://www.kroll.com/CMSPages/GetAzureFile.aspx?path=~%5Cmedia%5Cfiles%5Cintelligence-center%5Cglobal-fraud-report-2013-2014-english.pdf&hash=a84e800f9068acd3e3ce664e95666772c22ccf4547183e750ecf984b2041343d) that “All businesses are confronted with the risk of fraud. How they respond – the nature of their approach to prevention, detection, investigation and disclosure will separate those who manage through the issues from those who suffer significant loss.” An effective anti-fraud strategy has four main components:

### Prevention
One way to deal with fraud is to adopt methods that will decrease motive or restrict opportunity for potential fraudsters. The EMV technology is one example in preventing credit card fraud. Prevention include the introduction of policies, procedures and controls as well as activities such as training and fraud awareness to stop fraud from occurring.

In healthcare older adults can be at risk for scams or vulnerable to abuse of medical resources. The Healthcare Fraud Prevention Partnership (HFPP), a voluntary, public-private partnership between the federal government, state agencies, law enforcement and private health insurance plans to share information to identify and reduce fraud, waste and abuse (FWA). There are more that the public payers or private payers can do to prevent FWA. Dun & Bratsteets suggests more rigorous provider registration processes, or a verification process with third-party data or means to track the relationships between businesses and individuals, 
to prevent a bad provider even getting on-board. 

Unfortunately the fraud prevention techniques, while worth investing in, cannot provide 100% protection. It is difficult to remove all opportunities for fraud.

### Detection
Detection is the process of identifying if a new application or a transaction is fraudulent or genuine. A fraud detection system typically filter applications or transactions by checking some essential conditions (e.g. valid PII, not an duplication, etc) and then scored by a predictive model. The predictive model scores each application or transaction with high or low risk of fraud and escalates those with high risk for the review of investigators. The feedbacks from the investigators will be used to improve the model. 

A predictive model can be built upon experts’ rules, i.e. rules based on knowledge from fraud experts. Alternatively, with Machine Learning (ML) techniques can efficiently discover fraudulent patterns.



 A Fraud Detection System (FDS) should not only detect fraud cases efficiently, but also be cost-effective in the sense that the cost invested in the system should not be higher than the loss due to frauds.

### Response
An organization should have a response plan in conjunction with fraud detection. The plan defines the appropriate levels of authority and responsibilities, and reporting lines in the event of a suspected fraud. It acts as a checklist of actions and a guide to follow in the event of fraud being suspected. The plan is designed to:
* minimize and recover losses;
* prevent further loss;
* collect evidence necessary for criminal, civil and/or disciplinary action;
* determine when to contact the law enforcement;
* assign responsibility for investigating the incident;
* review the reasons for the event, the measures to prevent a recurrence and strengthens future responses to fraud.

### Deterrance

While prevention identifies and stops existing fraud, deterrence attempts to eliminate factors that may cause fraud. Fraud deterrence attacks the root causes of fraud. Deterrence improves organizational procedures, both short term procedural changes and long-term cultural changes, to reduce or eliminate the causal factors of fraud. As Cendrowski et al. summarized, "...it looks at what could happen in the future given the process definitions in place, and the people operating that process. Deterrence is a preventive measure – reducing input factors."

<ol>
  <li>Identify all the areas where fraud is expected to occur and the possible types of fraud.</li>
  <li>Measure the risk of fraud and the overall exposure to the organization. Prioritize the risks based on fraud.</li>
  <li>Apply machine learning in detecting frauds early.</li>
  <li>Establish risk assessment and decide where to pay closer attention.</li>
  <li>Monitor the activity with dashboard and data visualization. Communicate it throughout the organization so that employees in the organization are aware about the happening in the organization.</li>

  <li>Fix the broken controls by enhancing the procedures and systems.</li>
