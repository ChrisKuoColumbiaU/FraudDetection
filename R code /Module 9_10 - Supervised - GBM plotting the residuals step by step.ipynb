### GBM

```r
#install.packages("gbm")
library(gbm)
library(dplyr)

randomSeed = 1337
set.seed(randomSeed)

#data(iris)
iris2 <-iris %>% mutate(response=ifelse(Species=='setosa',1,0))
#iris2 <- iris2[sample(nrow(iris2)),]
head(iris2)

#dat <- matrix(iris2$response,gbmPredictions,residual) # make data
x=rep(1:length(iris2$response))
plot(x,iris2$response,type="o", col="blue", pch="o", lty=1, ylim=c(-2,2) )
```

##### n=1 tree

```r
gbmModel = gbm(formula = response ~ Sepal.Length+Sepal.Width+Petal.Length+Petal.Width,
               distribution = "bernoulli",
               data = iris2,
               n.trees = 1,
               shrinkage = 0.5)
gbmPredictions = predict(object = gbmModel,
                              newdata = iris2,
                              n.trees = 1,
                              type = "response")
residual = iris2$response - gbmPredictions

x=rep(1:length(iris2$response))
plot(x,iris2$response,type="o", col="blue", pch="o", lty=1, ylim=c(-2,2) )
# Add second curve to the same plot by calling points() and lines()
lines(x, gbmPredictions, col="red",lty=2,pch="+")
lines(x, residual, col="green",lty=2,pch="*")
legend(-2,-1,legend=c("Response","Predictions","Residuals"), col=c("blue","red","green"),
                                   pch=c("o","+","*"),lty=c(1,2,3))
```
