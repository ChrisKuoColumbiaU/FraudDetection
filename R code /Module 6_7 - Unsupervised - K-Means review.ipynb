## K-Means

In this notebook we will look at two examples. The first example is the classic iris dataset. We will first apply K-Means to cluster the iris dataset, then plot on a 2-dimensional graph. The second example is a dataset of 12 variables. In both examples we will use principal component analysis to reduce to 2-dimensions to plot on a 2-dimentional graph.

Learning objectives:
* K-Means
* Hyperparameter experimenting
* Dimension reduction 

Two examples:
* dataset 1: iris
* dataset 2: wine from library(rattle.data)

```r
#install.packages("rattle.data")
library(datasets)
library(rattle.data)
```

#### Example 1: iris

```r
head(iris)
table(iris$Species)
```

```r
library(ggplot2)
ggplot(iris, aes(Petal.Length, Petal.Width, color = Species)) + geom_point()
```

```r
set.seed(20)
irisCluster <- kmeans(iris[, 3:4], 3, nstart = 20) #This means that R will try 20 different random starting assignments and then select the one with the lowest within cluster variation.
irisCluster
```

```r
iris_PCA <- prcomp(iris[,1:4])
head(iris_PCA)
iris_PCA2 <-iris_PCA$x[,1:2]
```

```r
# Plot DBSCAN results
plot(iris_PCA2, main = "iris_PCA identified by K-means", frame = FALSE)
```

#### Example 2: Wine

```r
head(wine)
table(wine$Type)
```

```r
wn <-wine[,-1]
head(wn)
```

```r
wine_PCA <- prcomp(wn)
head(wine_PCA)
wine_PCA2 <-wine_PCA$x[,1:2]
```

```r
# Compute DBSCAN using fpc package
set.seed(1)
modl <- fpc::dbscan(wn, eps = 50, MinPts = 10)
plot(wine_PCA2, main = "Wine Type identified by DBSCAN", type="p", col=modl$cluster, frame = FALSE)
```
